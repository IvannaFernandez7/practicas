```python
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df_lluvias = pd.read_csv('../content/sample_data/pune_1965_to_2002.csv')
df_lluvias.head()
```

```python
df_lluvias.index = df_lluvias['Year']
df_lluvias = df_lluvias.drop('Year', axis='columns')


```

En el paso anterior por conveniencia pusimos a Year como indice de la tabla y lo borramos de las columnas

```python
df_lluvias.head()
```

```python
x = df_lluvias.index
y = df_lluvias.sum(axis='columns')
```

```python
ig, ax = plt.subplots(figsize=(12, 4))
ax.plot(x, y, label='Precipitaciones acumuladas')

ax.set_xlabel('Año')  				
ax.set_ylabel('Precipitación acumulada (mm.)')
ax.set_title('Precipitaciones acumuladas según el año')
ax.legend() 
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(x, y, label='Precipitaciones acumuladas')
fig.savefig("precipitaciones_año.pdf")
```
